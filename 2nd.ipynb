```python

import torch
import torchvision
from torchvision.models.detection import fasterrcnn_resnet50_fpn_v2
from torch.utils.data import Dataset, DataLoader
import os
from PIL import Image
import time
import datetime
from collections import defaultdict, deque
import numpy as np
from tqdm import tqdm

class SmoothedValue:
    def __init__(self, window_size=20):
        self.deque = deque(maxlen=window_size)
        self.total = 0.0
        self.count = 0

    def update(self, value):
        self.deque.append(value)
        self.count += 1
        self.total += value

    @property
    def median(self):
        d = torch.tensor(list(self.deque))
        return d.median().item()

    @property
    def avg(self):
        return sum(self.deque) / len(self.deque) if self.deque else 0.0

    @property
    def global_avg(self):
        return self.total / self.count if self.count > 0 else 0.0

class MetricLogger:
    def __init__(self):
        self.meters = defaultdict(SmoothedValue)
        self.iter_time = SmoothedValue()
        self.data_time = SmoothedValue()

    def update(self, **kwargs):
        for k, v in kwargs.items():
            if isinstance(v, torch.Tensor):
                v = v.item()
            self.meters[k].update(v)

    def __str__(self):
        loss_str = []
        for name, meter in self.meters.items():
            loss_str.append(f"{name}: {meter.median:.4f} ({meter.global_avg:.4f})")
        return "  ".join(loss_str)

class OptimizedKITTIDataset(Dataset):
    def __init__(self, root_dir, transforms=None, subset_fraction=0.2):
        self.root_dir = root_dir
        self.transforms = transforms
        self.image_dir = os.path.join(root_dir, '/root/.cache/kagglehub/datasets/klemenko/kitti-dataset/versions/1/data_object_image_2/training/image_2')
        self.label_dir = os.path.join(root_dir, '/root/.cache/kagglehub/datasets/klemenko/kitti-dataset/versions/1/data_object_label_2/training/label_2')

        if not os.path.exists(self.image_dir) or not os.path.exists(self.label_dir):
            raise ValueError(f"Directories not found at {self.image_dir} or {self.label_dir}")

        all_images = sorted([f for f in os.listdir(self.image_dir) if f.endswith('.png')])
        subset_size = int(len(all_images) * subset_fraction)
        self.image_files = all_images[:subset_size]

        self.classes = ['Car', 'Pedestrian', 'Cyclist']
        self.class_to_idx = {cls: idx + 1 for idx, cls in enumerate(self.classes)}

    def __len__(self):
        return len(self.image_files)

    def __getitem__(self, idx):
        img_path = os.path.join(self.image_dir, self.image_files[idx])
        label_path = os.path.join(self.label_dir, self.image_files[idx].replace('.png', '.txt'))

        img = Image.open(img_path).convert("RGB")

        boxes, labels = [], []
        with open(label_path) as f:
            for line in f:
                data = line.split()
                class_name = data[0]
                if class_name in self.class_to_idx:
                    labels.append(self.class_to_idx[class_name])
                    boxes.append([float(x) for x in data[4:8]])

        boxes = torch.as_tensor(boxes, dtype=torch.float32)
        labels = torch.as_tensor(labels, dtype=torch.int64)

        target = {
            "boxes": boxes,
            "labels": labels,
            "image_id": torch.tensor([idx]),
            "area": (boxes[:, 3] - boxes[:, 1]) * (boxes[:, 2] - boxes[:, 0]),
            "iscrowd": torch.zeros((len(boxes),), dtype=torch.int64)
        }

        if self.transforms:
            img = self.transforms(img)

        return img, target

def box_iou(boxes1, boxes2):
    area1 = (boxes1[:, 2] - boxes1[:, 0]) * (boxes1[:, 3] - boxes1[:, 1])
    area2 = (boxes2[:, 2] - boxes2[:, 0]) * (boxes2[:, 3] - boxes2[:, 1])

    lt = torch.max(boxes1[:, None, :2], boxes2[:, :2])
    rb = torch.min(boxes1[:, None, 2:], boxes2[:, 2:])

    wh = (rb - lt).clamp(min=0)
    inter = wh[:, :, 0] * wh[:, :, 1]

    union = area1[:, None] + area2 - inter
    return inter / union


def train_one_epoch(model, optimizer, data_loader, device, epoch, print_freq=10):
    model.train()
    metric_logger = MetricLogger()
    start_time = time.time()

    total_correct = 0
    total_objects = 0

    for batch_idx, (images, targets) in enumerate(data_loader):
        data_time = time.time() - start_time
        metric_logger.data_time.update(data_time)

        images = [image.to(device) for image in images]
        targets = [{k: v.to(device) for k, v in t.items()} for t in targets]

        # Forward pass and compute losses
        loss_dict = model(images, targets)
        losses = sum(loss for loss in loss_dict.values())

        # Compute accuracy
        model.eval()  # Switch to eval mode for predictions
        with torch.no_grad():
            predictions = model(images)
            model.train()  # Switch back to training mode

            # Calculate accuracy for this batch
            for pred, target in zip(predictions, targets):
                pred_boxes = pred['boxes']
                pred_scores = pred['scores']
                pred_labels = pred['labels']

                true_boxes = target['boxes']
                true_labels = target['labels']

                if len(pred_boxes) > 0 and len(true_boxes) > 0:
                    # Calculate IoU between predicted and true boxes
                    iou = box_iou(pred_boxes, true_boxes)
                    max_iou, matched_idx = iou.max(dim=1)

                    # Count correct predictions (IoU > 0.5 and correct class)
                    correct_class = pred_labels[max_iou > 0.5] == true_labels[matched_idx[max_iou > 0.5]]
                    total_correct += correct_class.sum().item()
                    total_objects += len(true_boxes)

        # Backward pass
        optimizer.zero_grad()
        losses.backward()
        optimizer.step()

        # Update metrics
        batch_accuracy = total_correct / total_objects if total_objects > 0 else 0
        metric_logger.update(loss=losses, accuracy=batch_accuracy, **loss_dict)

        if batch_idx % print_freq == 0:
            elapsed = time.time() - start_time
            eta_seconds = elapsed * (len(data_loader) - batch_idx) / (batch_idx + 1)
            eta = str(datetime.timedelta(seconds=int(eta_seconds)))

            print(f"Epoch: [{epoch}][{batch_idx}/{len(data_loader)}]  "
                  f"eta: {eta}  "
                  f"lr: {optimizer.param_groups[0]['lr']:.6f}  "
                  f"accuracy: {batch_accuracy:.4f}  "
                  f"{str(metric_logger)}")

        start_time = time.time()

    return {name: meter.global_avg for name, meter in metric_logger.meters.items()}

def evaluate(model, data_loader, device):
    model.eval()
    metric_logger = MetricLogger()

    with torch.no_grad():
        for images, targets in tqdm(data_loader, desc="Evaluation"):
            images = [image.to(device) for image in images]
            targets = [{k: v.to(device) for k, v in t.items()} for t in targets]

            outputs = model(images)

            # Calculate metrics
            for output, target in zip(outputs, targets):
                pred_boxes = output['boxes']
                pred_scores = output['scores']
                pred_labels = output['labels']

                true_boxes = target['boxes']
                true_labels = target['labels']

                if len(pred_boxes) > 0 and len(true_boxes) > 0:
                    iou = box_iou(pred_boxes, true_boxes)
                    max_iou, matched_idx = iou.max(dim=1)

                    # Calculate detection accuracy
                    correct_class = pred_labels[max_iou > 0.5] == true_labels[matched_idx[max_iou > 0.5]]
                    correct = correct_class.sum().item()
                    total = len(true_boxes)

                    metric_logger.update(
                        detection_accuracy=correct / total,
                        mean_iou=max_iou.mean().item()
                    )

    return {name: meter.global_avg for name, meter in metric_logger.meters.items()}

def main():
    # Configuration
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f"Using device: {device}")

    # Dataset setup
    transform = torchvision.transforms.Compose([
        torchvision.transforms.ToTensor(),
    ])

    # Hyperparameters
    batch_size = 3
    num_epochs = 10
    learning_rate = 0.001
    print_freq = 10

    # Initialize dataset and dataloaders
    dataset = OptimizedKITTIDataset(
        root_dir='/root/.cache/kagglehub/datasets/klemenko/kitti-dataset/versions/1',  # Update this path
        transforms=transform,
        subset_fraction=0.2
    )

    train_size = int(0.8 * len(dataset))
    val_size = len(dataset) - train_size
    train_dataset, val_dataset = torch.utils.data.random_split(dataset, [train_size, val_size])

    train_loader = DataLoader(
        train_dataset,
        batch_size=batch_size,
        shuffle=True,
        num_workers=2,
        collate_fn=lambda x: tuple(zip(*x))
    )

    val_loader = DataLoader(
        val_dataset,
        batch_size=batch_size,
        shuffle=False,
        num_workers=2,
        collate_fn=lambda x: tuple(zip(*x))
    )

    # Model initialization
    model = fasterrcnn_resnet50_fpn_v2(
        weights="DEFAULT",
        box_score_thresh=0.5,
        rpn_pre_nms_top_n_train=2000,
        rpn_post_nms_top_n_train=1000,
    )

    num_classes = len(dataset.classes) + 1
    model.roi_heads.box_predictor.cls_score = torch.nn.Linear(
        in_features=model.roi_heads.box_predictor.cls_score.in_features,
        out_features=num_classes
    )
    model.to(device)

    # Optimizer and scheduler
    params = [p for p in model.parameters() if p.requires_grad]
    optimizer = torch.optim.SGD(params, lr=learning_rate, momentum=0.9)
    scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=3, gamma=0.1)

    # Training loop
    best_accuracy = 0.0
    for epoch in range(num_epochs):
        print(f"\nEpoch {epoch+1}/{num_epochs}")

        # Train
        train_metrics = train_one_epoch(
            model, optimizer, train_loader, device, epoch, print_freq
        )

        # Evaluate
        eval_metrics = evaluate(model, val_loader, device)

        # Print metrics
        print("\nTraining metrics:")
        for k, v in train_metrics.items():
            print(f"{k}: {v:.4f}")

        print("\nValidation metrics:")
        for k, v in eval_metrics.items():
            print(f"{k}: {v:.4f}")

        # Save best model
        if eval_metrics['detection_accuracy'] > best_accuracy:
            best_accuracy = eval_metrics['detection_accuracy']
            torch.save({
                'epoch': epoch,
                'model_state_dict': model.state_dict(),
                'optimizer_state_dict': optimizer.state_dict(),
                'accuracy': best_accuracy,
            }, 'best_model.pth')

        scheduler.step()

    print("\nTraining completed!")
    print(f"Best validation accuracy: {best_accuracy:.4f}")

if __name__ == "__main__":
    main()
```
